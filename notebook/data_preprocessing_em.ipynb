# Merge the two datasets

```python
#drop offense End Date column from data2
data2 = data2.drop('Offense End Date', axis=1)
data2.head()
```

```python
data_test = data2.copy()
```

```python
# in data1 we have a column named 'Offense Start Date' that is has input format ex. '1/1/2009 0:00 AM'. We want to split this column into two columns one for Occur Date and one for Occur Time. In occur date place the '1/1/2009' and in occur time place '0:00 AM'
data_test[['Occur Date', 'Occur Time']] = data_test['Offense Start Date'].str.split(' ', 1, expand=True)
data_test.head()
```

```python
# Ensure 'Offense Start Date' is treated as a string
data2['Offense Start Date'] = data2['Offense Start Date'].astype(str)

# Now split 'Offense Start Date' into 'Occur Date' and 'Occur Time'
data2[['Occur Date', 'Occur Time']] = data2['Offense Start Date'].str.split(' ', n=1, expand=True)
data2.head()
```
